```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate true random normal data
np.random.seed(42)
data = np.random.normal(loc=26.5, scale=0.15, size=150)

plt.hist(data, bins=10, density=True, edgecolor='black')

# Add normal curve
mu = np.mean(data)
sigma = np.std(data)
x = np.linspace(min(data), max(data), 100)
plt.plot(x, norm.pdf(x, mu, sigma), linewidth=2)

plt.title("Histogram with Normal Curve")
plt.xlabel("Shoe Length (cm)")
plt.ylabel("Density")

plt.show()
```
